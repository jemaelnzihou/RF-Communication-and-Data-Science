```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.ticker as ticker
import math

# Set plot style
sns.set(style="whitegrid")

# Prepare data and plots for each concept from 0 to 5

# 0. Light vs. Electrical Communication
fiber_speed = 2e8  # meters per second
copper_speed = 1e8  # meters per second
distance = np.linspace(0, 1000, 100)  # in meters
fiber_time = distance / fiber_speed
copper_time = distance / copper_speed

plt.figure(figsize=(8, 5))
plt.plot(distance, fiber_time * 1e6, label='Optical Fiber (light)', lw=2)
plt.plot(distance, copper_time * 1e6, label='Copper Cable (electric)', lw=2)
plt.xlabel("Distance (meters)")
plt.ylabel("Latency (microseconds)")
plt.title("0. Latency Comparison: Light vs Electrical Communication")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### ðŸŽ¯ Why It Matters

###  4. RF Signal Strength over Distance (Friis Equation)

```
Pt = 1  # Transmitted power
Gt = 1  # Transmitter gain
Gr = 1  # Receiver gain
lambda_ = 0.125  # Wavelength in meters (2.4 GHz)
d = np.linspace(1, 100, 100)
L = 1
Pr = Pt * Gt * Gr * (lambda_ ** 2) / ((4 * np.pi * d) ** 2 * L)

plt.figure(figsize=(8, 4))
plt.plot(d, 10 * np.log10(Pr), label='Received Power (dB)', color='green')
plt.title("4. RF Signal Strength vs Distance (Friis Equation)")
plt.xlabel("Distance (m)")
plt.ylabel("Power Received (dB)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

ðŸ“¶ Plotted received RF power in dB over distance.


### Result

Shows rapid signal loss over distanceâ€”relevant for sensor range and data collection reliability.

### ðŸ“Š Insight

Signal strength (in dB) drops exponentially with distance, which is a key constraint in wireless data transmission.

At ~100 meters, signal strength has dropped by ~40 dB.

### ðŸŽ¯ Why It Matters

In smart factories or outdoor data-gathering systems (e.g., environmental, defense), RF sensors are used to collect data from hard-to-reach places.

That data eventually reaches data centers for processing and modeling.

Engineers must design with RF propagation limits in mindâ€”especially for IoT and smart infrastructure.

## Data Center Power Consumption vs Time

```
hours = np.arange(0, 24, 1)
power_usage = 5 + 2 * np.sin(2 * np.pi * hours / 24) + np.random.normal(0, 0.3, 24)

plt.figure(figsize=(8, 4))
plt.plot(hours, power_usage, marker='o', label='Power (kW)')
plt.title("5. Power Consumption in Data Center Over a Day")
plt.xlabel("Hour of Day")
plt.ylabel("Power Usage (kW)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Result:

### ðŸ“Š Insight

Data center power fluctuates with usage load and cooling demand.

### ðŸŽ¯ Why It Matters
